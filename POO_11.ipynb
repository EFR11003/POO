```python
import pandas as pd
import matplotlib.pyplot as plt
import math
import numpy as np
from pandas_datareader import data as pdr #Esta funcion descarga datos a partir de fuentes distintas a yfinance
import yfinance as yf
```

```python
yf.pdr_override()
```

```python
apple=pdr.get_data_yahoo("AAPL", start="2022-10-23", end= "2022-12-23")
apple
```

```python
#Crecimiento anual compuesto sobre fechas dadas
#Desv est y hacer sim
apple['Adj Close'][-1] #El argumento con menos 1 te da el ultimo valor de la serie
```

```python
crecimiento= apple['Adj Close'][1]/apple['Adj Close'][-1]-1
crecimiento #Esto es el HPR del ultimo dato al primer dato
```

```python
#Anualizar %
tiempo=(apple.index[-1]-apple.index[0]).days
tiempo #Esto me dice el numero de dias
```

```python
apple.describe()
```

```python
crecimeintocomp= crecimiento**(43)-1
crecimeintocomp
```

```python
#Volatilidad de los rendimeintos
volatilidad=apple["Adj Close"].pct_change().std()
volatilidad
```

```python
volatilidadtotal=volatilidad*math.sqrt(252) #Otra forma es 1/252
volatilidadtotal
```

```python
#Monte Carlo
#Vamos a generar porcentajes de rendimeinto diario
rendimeintosdiarios=np.random.normal(apple.pct_change().mean())
rendimeintosdiarios
```
